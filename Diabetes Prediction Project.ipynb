<div style="direction: rtl; unicode-bidi: embed; font-family: 'B Nazanin', sans-serif; font-size: 18px; line-height: 1.6;">
همانطور که ملاحظه می نمایید، میانگین و انحراف معیار شاخص های عددی در گروه های مختلف دیابتی با هم اختلاف دارند. حال برای اینکه بسنجیم آیا این اختلافات از لحاظ آماری معنا دار هستند یا خیر راه های مختلفی وجود دارد که در کد زیر از آزمون 
    <span style="font-family: 'Arial', sans-serif; font-weight: normal;">ANOVA</span>
    استفاده شده است که با کمک 
        <span style="font-family: 'Arial', sans-serif; font-weight: normal;">p-value</span>
        نشان میدهد میانگین های چند گروه آیا به صورت معنا دار اختلاف دارند یا خیر؟:
</div>


```python
from scipy import stats

x = ['BMI','Age','MentHlth','PhysHlth','GenHlth','Education','Income']

for i in x:
    group_0 = df[df['Diabetes_012'] == 0][i]
    group_1 = df[df['Diabetes_012'] == 1][i]
    group_2 = df[df['Diabetes_012'] == 2][i]
    #ANOVA Test:
    f_statistic, p_value = stats.f_oneway(group_0, group_1, group_2)
    print(f"P-value: {p_value}")
    if p_value < 0.05:
        print(i + " differences between 3 types of diabetes groups is significant!")
    else:
        print(i + " differences between 3 types of diabetes groups is not significant!")
```

```python
def calculate_proportions(df, feature, target):
    df['merged_target'] = df[target].replace({1: '1_or_2', 2: '1_or_2'})
    proportions = df.groupby('merged_target')[feature].value_counts(normalize=True).unstack().fillna(0)
    return proportions

#Binary features:
Binary_Features = ['HighBP', 'HighChol', 'CholCheck', 'Smoker', 'Stroke', 'HeartDiseaseorAttack', 
            'Fruits', 'Veggies', 'HvyAlcoholConsump', 'AnyHealthcare', 'NoDocbcCost', 
            'DiffWalk', 'Sex']


proportions = highbp_proportions = calculate_proportions(df, 'HighBP', 'Diabetes_012')

for i in Binary_Features:
    if i != 'HighBP':
        proportions = pd.concat([proportions,calculate_proportions(df, i, 'Diabetes_012')],axis = 1)
    
proportions

columns = []

for i in Binary_Features:
    columns.append((i,0))
    columns.append((i,1))
 
columns = pd.MultiIndex.from_tuples(
    columns,
    names=['Feature Name', 'Value']
)
proportions.columns = columns

pd.set_option('display.max_columns', None)

proportions
```

```python
x = proportions.iloc[1,range(1,26,2)] -  proportions.iloc[0,range(1,26,2)] 
x = np.abs(x)
x = x.sort_values(ascending=False)
x.index = x.index.map(lambda tup: tup[0])


plt.figure(figsize=(12, 8))
plt.plot(x.index, x.values, marker='o', linestyle='-', color='b')

plt.xlabel('Features', fontsize=14)
plt.ylabel('Percentage Increase (%)', fontsize=14)
plt.title('Percentage Increase in the Chance of Being in Diabetes and Pre-Diabetes Groups by Feature', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import chi2_contingency

def perform_chi2_test(df, feature, target):
    # Create a contingency table
    contingency_table = pd.crosstab(df[feature], df[target])
    
    # Perform the Chi-Square test
    chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)
    
    return chi2_stat, p_value

# Perform Chi-Square Test for each feature
results = {}

for feature in Binary_Features:
    chi2_stat, p_value = perform_chi2_test(df, feature, 'Diabetes_012')
    results[feature] = {'Chi2 Statistic': chi2_stat, 'p-value': p_value}

# Convert results to DataFrame for better readability
results_df = pd.DataFrame(results).T
print(results_df)

# Interpretation
alpha = 0.05
results_df['Significant'] = results_df['p-value'] < alpha
print(results_df)
```
